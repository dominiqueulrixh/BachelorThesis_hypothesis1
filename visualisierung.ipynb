```python
# 1. Setup
from model.housing_market_model import HousingMarketModel
import pandas as pd
import matplotlib.pyplot as plt

# 2. Simulation ausführen
model = HousingMarketModel()
for step in range(30):  # 30 Schritte simulieren
    model.step()

# 3. Daten auslesen
data = model.datacollector.get_model_vars_dataframe()

# 4. DataFrame anzeigen
data.tail()
```

```python
# 5. Verkaufszahlen über Zeit visualisieren

plt.figure(figsize=(10, 6))
plt.plot(data["Completed Sales"], label="Abgeschlossene Verkäufe")
plt.plot(data["Broker Sales"], label="Verkäufe über Broker", linestyle="--")
plt.xlabel("Zeitschritt")
plt.ylabel("Anzahl")
plt.title("Entwicklung der Verkäufe im Immobilienmarktmodell")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# 6. Marktaktivität anzeigen: aktive Käufer vs. verfügbare Immobilien

plt.figure(figsize=(10, 6))
plt.plot(data["Active Buyers"], label="Aktive Käufer")
plt.plot(data["Listed Properties"], label="Angebotene Immobilien")
plt.xlabel("Zeitschritt")
plt.ylabel("Anzahl")
plt.title("Marktaktivität im Modell: Nachfrage und Angebot")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# 7. Daten exportieren (optional)

data.to_csv("exportFiles/simulationsdaten.csv", index=False)
print("Daten erfolgreich als 'simulationsdaten.csv' exportiert.")
```
